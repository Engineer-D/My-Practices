## Comment generation

```python
def generate_comment(Question, Answer):
  '''
  Function that returns a generated comment based on question and response
  '''

  #Prompt
  prompt = \
  f"""
  Given
  Question: {Question}
  Answer: {Answer}
  Generate a comment to back up your answer
  """
  #comment Generation
  comment = openai.Completion.create(engine = 'text-davinci-003', prompt = prompt,
                                     max_tokens=50,
                                     temperature=0.7,
                                     top_p=1,
                                     n=1,
                                     stream=False)
  
  comment = comment.choices[0].text.strip()

  return comment
```

```python
import pandas as pd
```

```python
df = pd.read_csv("/content/QuestionTest.csv", delimiter="\t", header = None)
```

```python
df.columns = ["Question", "Answer"]
df['embeddings'] = 0
df['meta'] = 0
```

```python
df.head()
```

```python
from tqdm.auto import tqdm
```

```python
count = 0  # we'll use the count to create unique IDs
batch_size = 32  # process everything in batches of 32
for i in tqdm(range(0, len(df['Question']), batch_size)):
    # set end position of batch
    i_end = min(i+batch_size, len(df['Question']))
    # get batch of lines and IDs
    lines_batch = df['Question'][i: i_end]
    ids_batch = [str(n) for n in range(i, i_end)]
    # create embeddings
    res = openai.Embedding.create(input=list(lines_batch), engine=MODEL)
    embeds = [record['embedding'] for record in res['data']]
    df["embeddings"][i:i_end] = embeds
    # prep metadata and upsert batch
    meta = [{'text': line} for line in lines_batch]
    df["meta"][i:i_end] = meta
    # to_upsert = zip(ids_batch, embeds, meta)
    # upsert to Pinecone
    # index.upsert(vectors=list(to_upsert))
```

```python
df[['embeddings', 'meta']].head()
```

```python
query = list(df['Question'][[2,34,52,12,77,3]])

xq = openai.Embedding.create(input=query, engine=MODEL) #['data'][0]['embedding']
```

####Identical Questions

```python
for i in tqdm(range(len(xq['data']))):
  res = index.query([xq['data'][i]['embedding']], top_k=3, include_metadata=True)
  print(f"Question: {query[i]}")
  for match in res['matches']:
    print(f"{match['score']:.2f}: {match['metadata']['text']}")
    answer = df[df["Question"] == match['metadata']['text']]["Answer"].values
    print(f'Answer: {answer}')
  print()
```
